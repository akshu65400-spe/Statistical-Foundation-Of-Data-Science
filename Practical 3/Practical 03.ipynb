Question 1. Identify all duplicate cases using prof. Using all observations, find the average and standard deviation for age. Repeat the analysis by first filtering the data set to include one observation for each instructor with a total number of observations restricted to 94.

```python
import pandas as pd

data = pd.read_csv("data.csv")

duplicate_profs = data[data.duplicated(subset='prof', keep=False)].sort_values('prof')
print("Duplicate Records based on 'prof':")
print(duplicate_profs[['prof', 'age']].head(10))
print(f"\nTotal Duplicates Found: {len(duplicate_profs)}")

age_mean_all = data['age'].mean()
age_std_all = data['age'].std()
print(f"\nUsing All Observations -> Mean Age: {age_mean_all:.2f}, Std Dev: {age_std_all:.2f}")

data_unique = data.drop_duplicates(subset='prof', keep='first')

data_94 = data_unique.head(94)

age_mean_94 = data_94['age'].mean()
age_std_94 = data_94['age'].std()
print(f"Using One per Instructor (Restricted to 94) -> Mean Age: {age_mean_94:.2f}, Std Dev: {age_std_94:.2f}")
```

Question 2. Using a bar chart, demonstrate if instructors teaching lower-division courses receive higher average teaching evaluations.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

avg_rating_by_division = df.groupby('division')['rating'].mean()

plt.figure(figsize=(6, 4))
avg_rating_by_division.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Average Teaching Evaluation by Course Division')
plt.xlabel('Course Division')
plt.ylabel('Average Teaching Evaluation (Rating)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print(avg_rating_by_division)
```

Question 3. Plot the relationship between age and teaching evaluation scores.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

plt.figure(figsize=(7,5))
plt.scatter(df['age'], df['rating'], color='teal', alpha=0.6, edgecolor='black')

plt.title('Relationship between Age and Teaching Evaluation Scores')
plt.xlabel('Instructor Age')
plt.ylabel('Teaching Evaluation Score (Rating)')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

Question 4. Using gender-differentiated scatter plots, plot the relationship between age and teaching evaluation scores.

```python
import seaborn as sns
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='rating', hue='gender', data=df, palette={'Male':'purple','Female':'pink'})
plt.title("Age vs Teaching Evaluation by Gender")
plt.xlabel("Age")
plt.ylabel("Evaluation Score")
plt.legend(title='Gender')
plt.show()
```

Question 5. Create a scatterplot of age and evaluation scores, differentiated by gender and tenure

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.scatterplot(x='age', y='rating', hue='gender', style='tenure', data=df,
                palette={'Male':'purple','Female':'pink'}, s=100)
plt.title("Age vs Teaching Evaluation by Gender and Tenure")
plt.xlabel("Age")
plt.ylabel("Evaluation Score")
plt.legend(title='Gender/Tenure')
plt.show()
```
